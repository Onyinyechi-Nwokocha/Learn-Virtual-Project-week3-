# 4. Create BarCharts from the Categorcial Columns

```python
plt.style.use('ggplot')
data = {'Yes': 5, 'No' : 7}
names = list(data.keys())
values = list(data.values())
plt.bar(names,values,color='#55f6c1')
plt.title('Responses on Fear of walking alone')
plt.xlabel('Fear of Walking alone(Responses)')
plt.ylabel('No. of Responses')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 6. Explore how to Create Piecharts

```python
plt.style.use('fivethirtyeight')
slices = [59219, 55466, 47544, 36443, 35917]
#how many respondents said they know a particular language
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
#languages knows by respondents
colors = ['#d9184b','#f007b0', '#ed0d25', '#131372', '#6725ab']
explode = [0.1,0,0,0,0] #explode one part of the pie chart

plt.pie(slices,labels=labels,colors=colors,explode=explode,shadow=True,startangle=90,wedgeprops = {'edgecolor' : 'black'},autopct ='%1.1f%%')

plt.title("Responses on Programming Languages")
plt.tight_layout() #Default padding to make it look nice
plt.show()
```

# 7. Create a Dataframe for Reported Covid-19 cases in Nigeria, numbers vs state, choose wisely the best plot to visualize it

```python
df_Covid = pd.read_excel(r'C:\Users\Onyin\Desktop\Reported COVID-19 Cases.xlsx')
```

```python
new_index = (df_Covid.index + 1)
df_Covid.set_index(new_index, inplace=True)
```

```python
df_Covid 
```

```python
y = df_Covid.States_Affected
x = df_Covid.Lab_Cases
plt.figure(figsize=(15,15))
plt.style.use('seaborn-white')
plt.barh(y,x,color='#ff7e47',align ='center')
plt.title('No. of Lab confirmed COVID-19 Cases',fontweight='bold',fontsize=20)
plt.xticks(fontsize=16,rotation='horizontal')
plt.yticks(fontsize=16,rotation='horizontal')
for i, v in enumerate(x):
    plt.text(v + 3, i + .25, str(v), color='#000000', fontweight='bold',fontsize=20)
plt.grid(False)
plt.gca().invert_yaxis()
plt.ylabel('States Affected',fontweight='bold',fontsize=20)
plt.xlabel('No. of Cases',fontweight='bold',fontsize=20)
plt.show()
```

```python
y = df_Covid.States_Affected
x = df_Covid.On_Admission
plt.figure(figsize=(15,15))
plt.style.use('ggplot')
plt.barh(y,x,color='b',align ='center')
plt.title('No. of COVID-19 Cases on Admission',fontsize=20,fontweight='bold')
plt.xticks(fontsize=16,rotation='horizontal')
plt.yticks(fontsize=16,rotation='horizontal')
for i, v in enumerate(x):
    plt.text(v + 3, i + .25, str(v), color='#000000', fontweight='bold',fontsize=20)
plt.grid(False)
plt.gca().invert_yaxis()
plt.ylabel('States Affected',fontsize=20,fontweight='bold')
plt.xlabel('No of Cases',fontsize=20,fontweight='bold')
plt.show()
```

```python
y = df_Covid.States_Affected
x = df_Covid.No_Discharged
plt.figure(figsize=(15,15))
plt.style.use('ggplot')
plt.barh(y,x,color='#3BFB00',align ='center')
plt.title('No. of COVID-19 Cases Discharged',fontsize=20,fontweight='bold')
plt.xticks(fontsize=16,rotation='horizontal')
plt.yticks(fontsize=16,rotation='horizontal')
for i, v in enumerate(x):
    plt.text(v + 3, i + .25, str(v), color='#000000', fontweight='bold',fontsize=20)
plt.grid(False)
plt.gca().invert_yaxis()
plt.ylabel('States Affected',fontsize=20,fontweight='bold')
plt.xlabel('No of Cases',fontsize=20,fontweight='bold')
plt.show()
```

```python
y = df_Covid.States_Affected
x = df_Covid.No_of_Deaths
plt.figure(figsize=(15,15))
plt.style.use('ggplot')
plt.barh(y,x,color='r',align ='center')
plt.title('No. of COVID-19 Deaths',fontsize=20,fontweight='bold')
plt.xticks(fontsize=16,rotation='horizontal')
plt.yticks(fontsize=16,rotation='horizontal')
plt.gca().invert_yaxis()
for index,value in enumerate(x):
    plt.text(value,index,str(value),fontsize=22)
plt.grid(False)
plt.ylabel('States Affected',fontsize=20,fontweight='bold')
plt.xlabel('No of Cases',fontsize=20,fontweight='bold')
plt.show()
```
